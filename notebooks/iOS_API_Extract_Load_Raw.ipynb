```python
# prompt: Using dataframe df_apps: loop through the first 3 rows and print app_store_url

url = "https://ios-app-store-api1.p.rapidapi.com/details.php?url=https%3A%2F%2Fapps.apple.com%2Fus%2Fapp%2Fgoogle%2Fid284815942"

headers = {
    "x-rapidapi-key": "b173500eacmshf746569418243a8p147f18jsnad83cb36e882",
    "x-rapidapi-host": "ios-app-store-api1.p.rapidapi.com"
}

app_details = {
  'app_name': [],
  'app_store_url': [],
  'description': [],
  'ratingValue': [],
  'reviewCount': [],
  'category' : []
}

# Loop through the first 3 rows of the dataframe
for index in range(38):
    # Access the 'app_store_url' value for each row
    app_name = df_apps['app_name'].iloc[index]
    app_details['app_name'].append(app_name)

    app_store_url = df_apps['app_store_url'].iloc[index]
    app_details['app_store_url'].append(app_store_url)

    params = {"url":app_store_url}

    response = requests.get(url, headers=headers, params=params)

    # Print the app store URL
    print(f'app_name: {app_name} - app_store_url: {app_store_url}')

    # Assign the response data to the 'data' variable
    data = response.json()

    # Check if the keys exist before accessing them
    if 'description' in data[0]:
        description = data[0]['description']
        app_details['description'].append(description)
    else:
        app_details['description'].append(None)  # Or any other default value

    if 'rating' in data[0] and 'ratingValue' in data[0]['rating']:
        ratingValue = data[0]['rating']['ratingValue']
        app_details['ratingValue'].append(ratingValue)
    else:
        app_details['ratingValue'].append(None)

    if 'rating' in data[0] and 'reviewCount' in data[0]['rating']:
        reviewCount = data[0]['rating']['reviewCount']
        app_details['reviewCount'].append(reviewCount)
    else:
        app_details['reviewCount'].append(None)

    if 'offers' in data[0] and 'category' in data[0]['offers']:
        category = data[0]['offers']['category']
        app_details['category'].append(category)
    else:
        app_details['category'].append(None)

    #print(f'response: {response.json()}')
```
